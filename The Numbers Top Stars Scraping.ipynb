### Acting Credits

```python
# Driver
service = Service(executable_path=r'C:\chromedriver')
driver = webdriver.Chrome(service=service)
driver.get('https://www.the-numbers.com/box-office-star-records/worldwide/lifetime-acting/top-grossing-leading-stars')

acting_credit_data = []


# for i in range(1,130):
for i in range(1,2):
    try:
        page_ind  = i%6 + 2
        # for ind_star in range(1,101):
        for ind_star in range(1,2):
            star_link = driver.find_element(By.XPATH , f"""//*[@id="page_filling_chart"]/center/table/tbody/tr[{ind_star}]/td[2]/b/a""")
            star_link.click()
            
            # click Acting Credits
            actingcredits_butt = driver.find_element(By.XPATH , f"""//*[@id="a_acting"]""")
            actingcredits_butt.click()
            
            star_name = driver.find_element(By.XPATH , f"""//*[@id="main"]/div/h1""").text

            ac_table = driver.find_element(By.XPATH , f"""//*[@id="all_acting_credits"]/tbody""")
            # for tr in ac_table.find_elements(By.XPATH , './/tr'):
            for ind,tr in enumerate(ac_table.find_elements(By.XPATH , './/tr') ,start= 1):
                row = [item.text for item in tr.find_elements(By.XPATH , './/td')]
                movie_name = driver.find_element(By.XPATH , f'''//*[@id="all_acting_credits"]/tbody/tr[{ind}]/td[2]/b/a''').get_attribute('href')
                row = [star_name] + row + [movie_name]
                acting_credit_data.append(row)
            driver.back()
            driver.back()
        next_page = driver.find_element(By.XPATH , f"""//*[@id="page_filling_chart"]/center/div/a[{page_ind}]""")
        next_page.click()
    except:
        print(i)


# ac_cols_name = driver.find_element(By.XPATH,'//*[@id="all_acting_credits"]/thead/tr').find_elements(By.XPATH,'.//th')
# ac_cols_name_list = [i.text.replace('\n',' ') for i in ac_cols_name]
# ac_cols_name_list = ['Name'] + ac_cols_name_list + ['href']

ac_cols_name_list = ['Name',
'Release Date',
 'Title',
 'Role',
 'Domestic Box Office',
 'International Box Office',
 'Worldwide Box Office',
 'href']

ActingCredits = pd.DataFrame(acting_credit_data,columns= ac_cols_name_list)
ActingCredits.to_csv('all_acting_credits.csv',sep = '|')
display(ActingCredits.head())
```

### Star Information

```python
# Driver
service = Service(executable_path=r'C:\chromedriver')
driver = webdriver.Chrome(service=service)
driver.get('https://www.the-numbers.com/box-office-star-records/worldwide/lifetime-acting/top-grossing-leading-stars')

star_info_data = []

try:
    # for i in range(1,130):
    for i in range(1,2):
        
        page_ind  = i%6 + 2
        # for ind_star in range(1,101):
        for ind_star in range(1,30):
            star_link = driver.find_element(By.XPATH , f"""//*[@id="page_filling_chart"]/center/table/tbody/tr[{ind_star}]/td[2]/b/a""")
            star_link.click()
            
            star_tmp_list = [tr.text for tr in driver.find_element(By.XPATH,'//*[@id="col2mid"]/table/tbody').find_elements_by_tag_name("tr")]
            try:
                star_name = driver.find_element(By.XPATH , f"""//*[@id="main"]/div/h1""").text
            except : star_name= ''

            try:
                info_bestknown = driver.find_element(By.XPATH,"""//*[@id="col2mid"]/table/tbody/tr[1]/td/b""").text
                # print(info_bestknown)
            except:
                info_bestknown=''

            try:
                prodcoll = [i for i in  star_tmp_list if 'Most productive collaborators:' in i ][0]
                # print(prodcoll)
                prodcoll = [i.strip()  for i in prodcoll.replace('Most productive collaborators: ','').split(',')]
                # print( prodcoll  )
            except:
                prodcoll=''
                
            try:
                born = born = [i for i in  star_tmp_list if 'Born:' in i ][0]
                # print(born.replace('Born: ',''))
            except:
                born=''
                
            try:
                star_pic = driver.find_element(By.XPATH,"""//*[@id="person_img"]/img""").get_attribute('src')
                # print(star_pic)
            except:
                star_pic=''
            
            star_info_data.append([star_name,born,info_bestknown,prodcoll,star_pic])
            driver.back()

        next_page = driver.find_element(By.XPATH , f"""//*[@id="page_filling_chart"]/center/div/a[{page_ind}]""")
        next_page.click()
    
except:
    print(i)


star_info_df = pd.DataFrame(star_info_data)
star_info_df.columns = ['name','born','best_known_as','most_productive_collaborators','pic']
# star_info_df.to_csv('star_info_data.csv',sep = '|')
display(star_info_df.head())
print(star_info_df.shape)
```
